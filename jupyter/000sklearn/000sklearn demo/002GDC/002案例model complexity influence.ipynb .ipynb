```python
print(__doc__)
#输出文件开头注释的内容

#Author: dpslifeclass
#Linsence: BSD walkingPotato

import time;
import numpy as np;
import matplotlib.pyplot as plt;   
from mpl_toolkits.axes_grid1.parasite_axes import host_subplot;
from mpl_toolkits.axisartist.axislines import Axes;
from scipy.sparse.csr import csr_matrix;#(csr:Compressed Sparse Row marix压缩稀疏矩阵)

from sklearn import datasets;#数据集
from sklearn.utils import shuffle;#打乱数据
from sklearn.metrics import mean_squared_error;#均方误差
from sklearn.svm.classes import NuSVR;#支持向量机回归的一种算法
from sklearn.ensemble.gradient_boosting import GradientBoostingRegressor;#支持向量机模型融合
from sklearn.linear_model.stochastic_gradient import SGDClassifier;#SGD分类器
from sklearn.metrics import hamming_loss;#汉明损失

# #############################################################################
# Routines

# Initialize random generator
np.random.seed(0);#生成相同的随机数

#生成数据
def generate_data(case,sparse=False):
    """Generate regression/classification data."""
    bunch = None;
    if case == 'regression':
        bunch = datasets.load_boston();
    elif case == 'classification':
        bunch = datasets.fetch_20newsgroups_vectorized(subset='all');
    X,y = shuffle(bunch.data,bunch.target);#打乱数据
    print(type(X),type(y));#<class 'numpy.ndarray'> <class 'numpy.ndarray'>
    
    #根据偏移量确定训练集和测试集
    offset = int(X.shape[0] * 0.8);
    X_train,y_train = X[:offset],y[:offset];
    X_test,y_test = X[offset:],y[offset:];
    
    #如果是稀疏矩阵
    if sparse:
        X_train = csr_matrix(X_train);
        X_test = csr_matrix(X_test);
    else :
        X_train = np.array(X_train);
        X_test = np.array(X_test);
        
    y_test = np.array(y_test);
    y_train = np.array(y_train);
    
    data = {'X_train':X_train,"X_test":X_test,'y_train': y_train,
            'y_test': y_test};
    print(X_train.shape);
    print(X_test.shape);
    print(y_train.shape);
    print(y_test.shape);
    return data;

#测试生成数据方法
#generate_data('regression',False);

#计算基准影响，包含预测的性能，预测时间，计算复杂度
def benchmark_influence(conf):
    """
         Benchmark influence of :changing_param: on both MSE and latency.
    """
    prediction_times = [];
    prediction_powers = [];
    complexities = [];
    print("changing_param=%s" % (conf['changing_param']));
    for param_value in conf['changing_param_values']:
        print("changing_param_values: %f"% param_value);
        conf['tuned_params'][conf['changing_param']] = param_value;
        estimator = conf['estimator'](**conf['tuned_params']);#双星号表示参数是dict字典类型,estimator 分类器
        print("Benchmarking %s" % estimator);
        estimator.fit(conf['data']['X_train'],conf['data']['y_train']);#根据训练集和测试集训练模型
        conf['postfit_hook'](estimator);
        
        #计算复杂度
        complexity = conf['complexity_computer'](estimator);
        complexities.append(complexity);
        
        start_time = time.time();
        #开始预测，计算每个样本的平均预测时间
        for _ in range(conf['n_samples']):
            y_pred = estimator.predict(conf['data']['X_test']);
        elapsed_time = (time.time() - start_time) / float(conf['n_samples']);#过去的时间
        prediction_times.append(elapsed_time);
        
        #开始计算成绩
        pred_score = conf['prediction_performance_computer'](conf['data']['y_test'],y_pred);
        prediction_powers.append(pred_score);
        print("Complexity:%d | %s: %.4f | Pred Time: %fs\n"
              %(complexity,conf['prediction_performance_label'],pred_score,elapsed_time));
        
        print(prediction_powers[0:3]);
        print(prediction_times[0:3]);
        print(complexities[0:3]);
    return prediction_powers,prediction_times,complexities;
    
#画图
def plot_influence(conf,mse_values,prediction_times,complexities):
    """
         Plot influence of model complexity on both accuracy and latency.
         plt是画板
         figure是画布
         axes是绘图区
         画布在画板上，绘图区在画布上，一个画板可以有多个画布，一个画布可以有多个绘图区
         如果没有手动指定画布figure和绘图区axes or subplot，使用画板plt时会自动创建默认的画布和绘图区各一个。
    """
    plt.figure(figsize=(12,6));#划分子图
    host = host_subplot(111,axes_class=Axes);#创建一行一列的子图，放在第一个子图上
    plt.subplots_adjust(right=0.75);#用来调整子图的上下左右间距，值为百分比
    '''
        使用twinx是添加y轴的坐标轴

        使用twiny是添加x轴的坐标轴
    '''
    par1 = host.twinx();
               
    #设置x，y坐标轴标签
    host.set_xlabel('Model Complexity (%s)' % conf['complexity_label']);
    y1_label = conf['prediction_performance_label'];
    y2_label = "Time (s)";
    host.set_ylabel(y1_label);
    par1.set_ylabel(y2_label);
    
    print("绘图前数据，复杂度为%d,均方误差MSE为%f,预测时间为%f"%(complexities[0],mse_values[0],prediction_times[0]));
    #绘图
    p1, = host.plot(complexities,mse_values,'b-',label="prediction error");#参数值含义，横坐标，纵坐标，颜色，标签
    p2, = par1.plot(complexities,prediction_times,'r-',label="latency");
    
    #绘制图头
    host.legend(loc='upper right');
    host.axis["left"].label.set_color(p1.get_color());
    par1.axis["right"].label.set_color(p2.get_color());
    plt.title("Influence of Model Complexity - %s" % conf["estimator"].__name__);#使用分类器对象的类名作为标题title
               
    #显示图
    plt.show();
  
               
#统计非0西塔的个数
def _count_nonzero_coefficients(estimator):
    a = estimator.coef_.toarray();
    return np.count_nonzero(a);
        
# #############################################################################
# Main code
regression_data = generate_data('regression');
classification_data = generate_data('classification', sparse=True);
configurations = [
    {'estimator': SGDClassifier,
     'tuned_params': {'penalty': 'elasticnet', 'alpha': 0.001, 'loss':
                      'modified_huber', 'fit_intercept': True},
     'changing_param': 'l1_ratio',
     'changing_param_values': [0.25, 0.5, 0.75, 0.9],
     'complexity_label': 'non_zero coefficients',
     'complexity_computer': _count_nonzero_coefficients,
     'prediction_performance_computer': hamming_loss,
     'prediction_performance_label': 'Hamming Loss (Misclassification Ratio)',
     'postfit_hook': lambda x: x.sparsify(),
     'data': classification_data,
     'n_samples': 30},
    {'estimator': NuSVR,
     'tuned_params': {'C': 1e3, 'gamma': 2 ** -15},
     'changing_param': 'nu',
     'changing_param_values': [0.1, 0.25, 0.5, 0.75, 0.9],
     'complexity_label': 'n_support_vectors',
     'complexity_computer': lambda x: len(x.support_vectors_),
     'data': regression_data,
     'postfit_hook': lambda x: x,
     'prediction_performance_computer': mean_squared_error,
     'prediction_performance_label': 'MSE',
     'n_samples': 30},
    {'estimator': GradientBoostingRegressor,
     'tuned_params': {'loss': 'ls'},
     'changing_param': 'n_estimators',
     'changing_param_values': [10, 50, 100, 200, 500],
     'complexity_label': 'n_trees',
     'complexity_computer': lambda x: x.n_estimators,
     'data': regression_data,
     'postfit_hook': lambda x: x,
     'prediction_performance_computer': mean_squared_error,
     'prediction_performance_label': 'MSE',
     'n_samples': 30},
];

loopCount = 0;
for conf in configurations:
    loopCount += 1;
    print("****************************开始第%d次循环，机器学习算法%s******************************" % (loopCount,conf['estimator']));
    #计算预测值的性能，复杂度，执行时间
    prediction_performances, prediction_times, complexities = \
        benchmark_influence(conf);
    #画图
    plot_influence(conf, prediction_performances, prediction_times,
                   complexities);
    
print("**************************** 算法结束 ******************************");
```
